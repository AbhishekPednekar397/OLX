```python
# Encode values in Make column

lb = LabelEncoder()
df["Make_Encode"] = lb.fit_transform(df["Make"])
```

Machine Learning model accepts only numerical data. Hence by using Encoding Techniques we have to convert the string data in our dataset to numerical data.

We have used Label encoding Technique to encode the string values in Make column to numerical data. 
Seperate column Make_Encode column has been created to store the encoded values.

```python
# Encoding all the columns to convert string data into numerical data

lb = LabelEncoder()

feature = {}
for k in df.columns:
    if df[k].dtype == 'O':
        df[str(k)+"_Encode"] = lb.fit_transform(df[k])
        gt = {}
        ke = df[[k,str(k)+"_Encode"]].drop_duplicates()[str(k)+"_Encode"].values
        val = df[[k,str(k)+"_Encode"]].drop_duplicates()[k].values
        for i in range(len(ke)):
            gt[ke[i]] = val[i]
        feature[k] = gt
```
